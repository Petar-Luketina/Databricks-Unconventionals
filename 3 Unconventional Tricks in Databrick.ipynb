```
def dataframe_from_paste(data, headers=None, replace_headers=None):
  data_list = data.split('\n')[1:-1]
  if headers == True:
    headers = data_list.pop(0).split('\t')
  elif replace_headers:
    headers = list(map(lambda x: str(x), replace_headers))
    data_list.pop(0)
  temp_list = [row.split('\t') for row in data_list]
  return spark.createDataFrame(temp_list, schema=headers)

df = dataframe_from_paste(s, replace_headers=['first', 'last', 'employment'])
display(df)
```
